```python
import csv
import pickle
import numpy as np
import ast

alpha_file = '/dk/lcj_rag/pyserini-autofusion/experiments/runs/nq_train/run.dpr.nq-train.fusion.all-alpha.nq--1.k-100.res.test'

class0 = []
class1 = []
class2 = []
class3 = []
class4 = []
class5 = []
class6 = []
class_hardfail = []
class0_dist = []
class1_dist = []
class2_dist = []

with open(alpha_file, 'r') as f:
    reader = csv.reader(f, delimiter='\t')
    lines = list(reader)
    # (qid, hits, softmax_hits, weights)
    for i in range(len(lines)):
        hits = lines[i][1]
        hits = hits.replace('hits: ', '')
        hits = ast.literal_eval(hits)
        # print(i, hits)

        if sum(hits):
            if hits[0] == 0 and hits[1] == 0 and hits[-1] != 0:
                class0.append((i, hits))
            elif hits[0] == 0 and hits[1] != 0 and hits[-1] == 0:
                class1.append((i, hits))
            elif hits[0] != 0 and hits[1] == 0 and hits[-1] == 0:
                class2.append((i, hits))
            elif hits[0] != 0 and hits[1] != 0 and hits[-1] == 0:
                class3.append((i, hits))
            elif hits[0] == 0 and hits[1] != 0 and hits[-1] != 0:
                class4.append((i, hits))
            elif hits[0] != 0 and hits[1] == 0 and hits[-1] != 0:
                class5.append((i, hits))
            elif hits[0] != 0 and hits[1] != 0 and hits[-1] != 0:
                class6.append((i, hits))
        else:
            class_hardfail.append((i, hits))

print(f'class0: {len(class0)} | {len(class0) / len(lines)}')
print(f'class1: {len(class1)} | {len(class1) / len(lines)}')
print(f'class2: {len(class2)} | {len(class2) / len(lines)}')
print(f'class3: {len(class3)} | {len(class3) / len(lines)}')
print(f'class4: {len(class4)} | {len(class4) / len(lines)}')
print(f'class5: {len(class5)} | {len(class5) / len(lines)}')
print(f'class6: {len(class6)} | {len(class6) / len(lines)}')
print(f'fail: {len(class_hardfail)} | {len(class_hardfail) / len(lines)}')
print(len(class0) + len(class1) + len(class2) + len(class3) + len(class4) + len(class5) + len(class6) + len(class_hardfail))

# class0: [0, 0, 1] = 1362 | 0.01554812269546456
# class1: [0, 1, 0] = 81 | 0.0009246680898183769
# class2: [1, 0, 0] = 896 | 0.010228427265151429
# class3: [1, 1, 0] = 4051 | 0.0462448201463487
# class4: [0, 1, 1] = 14726 | 0.1681069418600669
# class5: [1, 0, 1] = 160 | 0.0018265048687770408
# class6: [1, 1, 1] = 48984 | 0.559184465576091
```
