```python
import os
```

```python
cmd0 = 'python -m pyserini.search.faiss \
  --index wikipedia-dpr-100w.dpr-multi \
  --topics dpr-wq-test \
  --encoded-queries dpr_multi-wq-test \
  --output runs/base_trec/run.dpr.wq-test.multi.bf.trec \
  --batch-size 108 --threads 36'
os.system(cmd0)
```

```python
cmd1 = 'python -m pyserini.eval.convert_trec_run_to_dpr_retrieval_run \
  --index wikipedia-dpr-100w \
  --topics dpr-wq-test \
  --input runs/base_trec/run.dpr.wq-test.multi.bf.trec \
  --output runs/base_trec/run.dpr.wq-test.multi.bf.json'
os.system(cmd1)
```

```python
cmd2 = 'python -m pyserini.eval.evaluate_dpr_retrieval \
  --retrieval runs/base_trec/run.dpr.wq-test.multi.bf.json \
  --topk 20 100'
os.system(cmd2)
```
