```python
import re
import json

json_path = '/dk/rag-dataset/webquestions.examples.train.json'
csv_path = '/dk/rag-dataset/topics.dpr.wq.train.txt'

with open(json_path, 'r') as f:
    data = f.read()

data = json.loads(data)

print(data[0])

# drop data['url']
data = [{k: v for k, v in d.items() if k != 'url'} for d in data]
print(data[0])

print(data[0]["answer"]) # (list (description "Jazmyn Bieber") (description "Jaxon Bieber"))

for item in data:
    for k, v in item.items():
      if k != 'answer':
          continue
      
      # if v.startswith('(list '):
      content = v[6:-1]
      matches = re.findall(r'\(description (.*?)\)', content)
      
      for i in range(len(matches)):
          if matches[i].startswith('"'):
              matches[i] = matches[i][1:-1]

      # print(len(matches), matches)
      
      list_content = ''
      for i in range(len(matches) - 1):
          list_content += f"'{matches[i]}'" + ', '
      list_content += f"'{matches[-1]}'"
      item[k] = '[' + list_content + ']'
      item[k] = matches
      
          

for item in data:
    print(item["answer"])

with open(csv_path, 'w') as f:
    for item in data:
        json.dump(item, f)
        f.write('\n')
```
